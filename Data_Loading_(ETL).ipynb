Question no. 9 Loading Strategy Selection

Assume this dataset represents daily sales data.

a) Should a full Load or Incremental Laod be used?

Answer: Incremental Load

b) Justify your choice.

1) Daily sales data is generated every day, so only new records need to be loaded.

2) Incremental loading is faster and reduces processing time.

3) It minimizes system and database resource usage.

4) It avoids reloading old data, reducing the risk of duplicate records.

5) It is the standard approach for regularly updated transactional data.
